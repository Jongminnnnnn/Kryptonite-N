```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

#  
X = np.load("Datasets/kryptonite-18-X.npy")
y = np.load("Datasets/kryptonite-18-y.npy")

#   
print("Feature shape:", X.shape)
print("Label shape:", y.shape)

#  
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# PCA  2D 
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# 2D 
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[y == 0, 0], X_pca[y == 0, 1], label='Class 0', alpha=0.5)
plt.scatter(X_pca[y == 1, 0], X_pca[y == 1, 1], label='Class 1', alpha=0.5)
plt.title("PCA Visualization of Kryptonite-18 Dataset")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.show()

#   
unique, counts = np.unique(y, return_counts=True)
print("Label Distribution:", dict(zip(unique, counts)))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
import umap
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

#  
X = np.load('Datasets/kryptonite-18-X.npy')
y = np.load('Datasets/kryptonite-18-y.npy')

#  
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# t-SNE 
tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X_scaled)

# UMAP 
umap_model = umap.UMAP(n_components=2, random_state=42)
X_umap = umap_model.fit_transform(X_scaled)

# t-SNE 
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.scatter(X_tsne[y == 0, 0], X_tsne[y == 0, 1], label="Class 0", alpha=0.5)
plt.scatter(X_tsne[y == 1, 0], X_tsne[y == 1, 1], label="Class 1", alpha=0.5)
plt.title("t-SNE Visualization of Kryptonite-18 Dataset")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.legend()

# UMAP 
plt.subplot(1, 2, 2)
plt.scatter(X_umap[y == 0, 0], X_umap[y == 0, 1], label="Class 0", alpha=0.5)
plt.scatter(X_umap[y == 1, 0], X_umap[y == 1, 1], label="Class 1", alpha=0.5)
plt.title("UMAP Visualization of Kryptonite-18 Dataset")
plt.xlabel("UMAP Component 1")
plt.ylabel("UMAP Component 2")
plt.legend()

plt.show()
```
